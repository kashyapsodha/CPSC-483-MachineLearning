# using the linear regression model from sklearn with mean squared error as the loss function

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

# using cross validation for spliting training and testing data.

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
X_train
```

```python
lr.fit(X_train,y_train)
```

```python
preds = lr.predict(X_test)
```

```python
np.asarray(y_test.iloc[0:])
```

```python
def avg_error(y_tes , preds):
    error = 0
    for i in range(len(y_tes)):
        error += np.abs(y_tes[0] - preds[0])
    print(error)
    return error/len(y_tes)
```

# checking the average error given by our model

```python
avg_error(np.asarray(y_test.iloc[0:]),preds)
```

```python
lr_new = LinearRegression()
```

```python
lr_new.fit(X_train.iloc[:,2:],y_train)
```

```python
new_preds = lr_new.predict(X_test.iloc[:,2:])
```
